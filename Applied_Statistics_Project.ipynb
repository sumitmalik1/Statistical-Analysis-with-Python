```python
columns_with_outliers = ['Income', 'Age', 'MntWines', 'MntFruits', 'MntMeatProducts']

def remove_outliers(data,columns):

    for column in columns:
        Q1 = data[column].quantile(0.20)
        Q3 = data[column].quantile(0.80)
        IQR = Q3  - Q1

#Defining the lower and upper bounds

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3  + 1.5 * IQR

        data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

    return data

```

```python
new_df = remove_outliers(df,columns_with_outliers)
new_df
```

## **Probability and Expected Value**

```python

import numpy as np
from scipy.stats import norm
```

```python
mean_recency = np.mean(new_df['Recency'])
std_recency = np.std(new_df['Recency'])

prob_less_than_20 = norm.cdf(20, loc = mean_recency, scale = std_recency)
prob_greater_than_20 = 1 - norm.cdf(21, loc = mean_recency, scale = std_recency)

print(f"Probability of recency less than 20: {prob_less_than_20:.3f}:")
print(f"Probability of recency less than 20: {prob_greater_than_20:.3f}:")


#expected value

expected_value = mean_recency
print(f"The expected value is: {expected_value:.2f}")
```

## **Probability and Expected Value**

```python
new_df
```

```python
from scipy.stats import binom

num_response = len(df['Response']) # Num of trails for a response
success_response = np.mean(df['Response']) # probability of success for Response
k1 = np.sum(df['Response'] == 1)  # No of success for getting response 1


num_Complain = len(df['Complain']) # Num of trails for a Complain
success_Complain = np.mean(df['Complain']) # probability of success for Complain
k2 = np.sum(df['Complain'] == 1) # No of success for getting complain 1

#probabilities of complain and responses are

prob_response = binom.pmf(k1,num_response,success_response)
prob_complain = binom.pmf(k2,num_Complain,success_Complain)

print(f"The probability of Response being 1 : {prob_response:.2f}")
print(f"The probability of Complain being 1 : {prob_complain:.2f}")

# Expected value

expected_response = num_response * success_response
expected_complain = num_Complain * success_Complain

print(f"The Expected value of Response is : {expected_response:.2f}")
print(f"The Expected value of Complain is: {expected_complain:.2f}")
```
